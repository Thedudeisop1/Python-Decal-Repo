```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.optimize as fit

#Exoplanet data from Habitable Worlds Catalog (HWC)
ExoData = pd.read_csv("/Users/joshuawong/Desktop/hwc.csv")
display(ExoData)


#Filtering out the data to get the desired rows and columns
ExoData = ExoData[["P_NAME","P_TYPE","P_TYPE_TEMP","S_DISTANCE","S_DISTANCE_ERROR_MIN","S_DISTANCE_ERROR_MAX","P_ESI"]]
display(ExoData)

#Dropped "NaN" Values
ExoData = ExoData.dropna()

#Specifed Type and Temperature of the Exo-Planets
ExoData = ExoData[ExoData["P_TYPE"].str.contains("Subterran|Terran|Superterran")]
ExoData = ExoData[ExoData["P_TYPE_TEMP"].str.contains("Hot|Warm")]
display(ExoData)


#Random Data and Plots to make sure Models can accurately fit them
xdata1 = [0,1,1.2,1.4,2,2.3,2.6,2.9,3,3.1,4]
ydata1 = [0,3,2,5,6,3,5,6,8,7,10]
xdata2 = [0,1,1.2,1.4,2,2.3,2.6,2.9,3,3.1,4]
ydata2 = [10,6,4,1.5,1,1.1,0.8,0.9,1,1.5,1.2]
xdata3 = [0,1,1.2,1.4,2,2.3,2.6,2.9,3,3.1,4]
ydata3 = [10,6,4,1.5,1,1.1,0.8,0.9,1,1.5,1.2]

fig,axes = plt.subplots(3)
fig.suptitle("Data Generation for Testing")

axes[0].scatter(xdata1,ydata1)
pplinear = [1, 0]
pp1, ccov1 = fit.curve_fit(linear,xdata1,ydata1,pplinear)
xx1 = np.linspace(np.min(xdata1),np.max(xdata1),100)
axes[0].plot(xx1, linear(xx1, pp1[0], pp1[1]), c="k")

axes[1].scatter(xdata2,ydata2)
ppexponential = [1, 1, 1, 1]
pp2, ccov2 = fit.curve_fit(exponential,xdata2,ydata2,ppexponential)
xx2 = np.linspace(np.min(xdata2),np.max(xdata2),100)
axes[1].plot(xx2, exponential(xx2, pp2[0], pp2[1], pp2[2], pp2[3]), c="k")

axes[2].scatter(xdata3,ydata3)
ppPowerLaw = [1, 1, 1]
pp3, ccov3 = fit.curve_fit(powerLaw,xdata3,ydata3,ppPowerLaw)
xx3 = np.linspace(np.min(xdata3),np.max(xdata3),100)
axes[2].plot(xx3, powerLaw(xx3, pp3[0], pp3[1], pp3[2]), c="k")


#Creating Scatterplots
fig0,ax0 = plt.subplots()
fig0.suptitle("Distributions of Data Points Between ESI and Planet Distance")
ax0.scatter(ExoData["S_DISTANCE"],ExoData["P_ESI"], s=1)
ax0.set_xlabel("Distance to Exo-Planet (Lightyears)")
ax0.set_ylabel("Earth Similarity Index (ESI)")


#Linear Fit and Residual Plots
def linear(x, a, b):
    return a * x + b

plinear = [-0.1/500, 0.4]
p1, cov1 = fit.curve_fit(linear,ExoData["S_DISTANCE"],ExoData["P_ESI"],plinear)

fig1,ax1 = plt.subplots()
fig1.suptitle("Linear Fit Between ESI and Planet Distance")
ax1.scatter(ExoData["S_DISTANCE"],ExoData["P_ESI"], s=1)
ax1.set_xlabel("Distance to Exo-Planet (Lightyears)")
ax1.set_ylabel("Earth Similarity Index (ESI)")
x1 = np.linspace(0,3500,900)
ax1.plot(x1,linear(x1, p1[0], p1[1]), c="r")

fig4,ax4 = plt.subplots()
fig4.suptitle("Linear Fit Residual Plot")
ax4.scatter(ExoData["S_DISTANCE"],ExoData["P_ESI"] - linear(ExoData["S_DISTANCE"], p1[0], p1[1]), s=1)
ax4.set_xlabel("Distance to Exo-Planet (Lightyears)")
ax4.set_ylabel("Residuals (ESI)")

P_errors1 = []
for i in range(len(cov1)):
    P_errors1.append(cov1[i][i] ** 0.5)
    
print('\nLinear Fit equation parameters and their errors:')
print('a = {0:1.5f} +/- {1:1.5f}'.format(p1[0], P_errors1[0]))
print('b = {0:1.5f} +/- {1:1.5f}'.format(p1[1], P_errors1[1]))

print('\nThe final equation for Linear Fit:')
print('{0:1.5f} x + {1:1.5f}'.format(p1[0], p1[1]))

#Analysis of Linear Fit
rSquared = 1 - (sum((ExoData["P_ESI"] - linear(ExoData["S_DISTANCE"], p1[0], p1[1]))**2)/sum((ExoData["P_ESI"] - np.mean(ExoData["P_ESI"])**2)))
print("\nThe R^2 value is")
print(rSquared)


#Exponential Fit and Residual Plots
def exponential(x, a, b, c, d):
    return a * np.exp(-b * x + c) + d

pexponential = [0.1, 0.4, 0.1, 0.1]
p2, cov2 = fit.curve_fit(exponential,ExoData["S_DISTANCE"],ExoData["P_ESI"],pexponential)

fig2,ax2 = plt.subplots()
fig2.suptitle("Exponential Fit Between ESI and Planet Distance")
ax2.scatter(ExoData["S_DISTANCE"],ExoData["P_ESI"], s=1)
ax2.set_xlabel("Distance to Exo-Planet (Lightyears)")
ax2.set_ylabel("Earth Similarity Index (ESI)")
x2 = np.linspace(0,3500,900)
ax2.plot(x2,exponential(x2, p2[0], p2[1], p2[2], p2[3]), c="gold")

fig5,ax5 = plt.subplots()
fig5.suptitle("Exponential Fit Residual Plot")
ax5.scatter(ExoData["S_DISTANCE"],ExoData["P_ESI"] - exponential(ExoData["S_DISTANCE"], p2[0], p2[1], p2[2], p2[3]), s=1)
ax5.set_xlabel("Distance to Exo-Planet (Lightyears)")
ax5.set_ylabel("Residuals (ESI)")

P_errors2 = []
for i in range(len(cov2)):
    P_errors2.append(cov2[i][i] ** 0.5)
    
print('\nExponential Fit equation parameters and their errors:')
print('a = {0:1.5f} +/- {1:1.5f}'.format(p2[0], P_errors2[0]))
print('b = {0:1.5f} +/- {1:1.5f}'.format(p2[1], P_errors2[1]))
print('c = {0:1.5f} +/- {1:1.5f}'.format(p2[2], P_errors2[2]))
print('d = {0:1.5f} +/- {1:1.5f}'.format(p2[3], P_errors2[3]))

print('\nThe final equation for Exponential Fit:')
print('{0:1.5f} exp(-{1:1.5f} x + {2:1.5f}) + {3:1.5f}'.format(p2[0], p2[1], p2[2], p2[3]))

#Analysis of Exponential Fit
chi2 = sum(((ExoData["P_ESI"] - exponential(ExoData["S_DISTANCE"], p2[0], p2[1], p2[2], p2[3])) / ExoData["S_DISTANCE_ERROR_MAX"])**2) 
rchi2 = chi2 / (len(ExoData["P_ESI"]) - 4)
print(f'\nReduced chi squared for Exponential Fit: {rchi2:.5f}')


#Power-Law Fit and Residual Plots
def powerLaw(x, a, b, c):
    return a * x ** -b + c

pPowerLaw = [0.1, 0.4, 0.1]
p3, cov3 = fit.curve_fit(powerLaw,ExoData["S_DISTANCE"],ExoData["P_ESI"],pPowerLaw)

fig3,ax3 = plt.subplots()
fig3.suptitle("Power-Law Fit Between ESI and Planet Distance")
ax3.scatter(ExoData["S_DISTANCE"],ExoData["P_ESI"], s=1)
ax3.set_xlabel("Distance to Exo-Planet (Lightyears)")
ax3.set_ylabel("Earth Similarity Index (ESI)")
x3 = np.linspace(0.05,3500,900)
ax3.plot(x3,powerLaw(x3, p3[0], p3[1], p3[2]), c="magenta")

fig6,ax6 = plt.subplots()
fig6.suptitle("Power-Law Fit Residual Plot")
ax6.scatter(ExoData["S_DISTANCE"],ExoData["P_ESI"] - powerLaw(ExoData["S_DISTANCE"], p3[0], p3[1], p3[2]), s=1)
ax6.set_xlabel("Distance to Exo-Planet (Lightyears)")
ax6.set_ylabel("Residuals (ESI)")

P_errors3 = []
for i in range(len(cov3)):
    P_errors3.append(cov3[i][i] ** 0.5)
    
print('\nPower-Law Fit equation parameters and their errors:')
print('a = {0:1.5f} +/- {1:1.5f}'.format(p3[0], P_errors3[0]))
print('b = {0:1.5f} +/- {1:1.5f}'.format(p3[1], P_errors3[1]))
print('c = {0:1.5f} +/- {1:1.5f}'.format(p3[2], P_errors3[2]))

print('\nThe final equation for the Power-Law Fit:')
print('{0:1.5f} x ^ (-{1:1.5f}) + {2:1.5f}'.format(p3[0], p3[1], p3[2]))

#Analysis of Power-Law Fit
chi3 = sum(((ExoData["P_ESI"] - powerLaw(ExoData["S_DISTANCE"], p3[0], p3[1], p3[2])) / ExoData["S_DISTANCE_ERROR_MAX"])**2) 
rchi3 = chi3 / (len(ExoData["P_ESI"]) - 3)
print(f'\nReduced chi squared for Power-Law Fit: {rchi3:.5f}')
```
