```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = sns.load_dataset("planets")
df.head(15)

methodList = []
for i in df.method:
    methodList.append(i)
orbitalPeriodList = []
for i in df.orbital_period:
    orbitalPeriodList.append(i)
massList = []
for i in df.mass:
    massList.append(i)

astrometryList = []
eclipseTimingVariationsList = []
imagingList = []
microlensingList = []
orbitalBrightnessModulationList = []
pulsarTimingList = []
pulsationTimingVariationsList = []
radialVelocityList = []
transitList = []
transitTimingVariationsList = []
for i in list(range(1035)):
    if methodList[i] == "Astrometry":
        astrometryList.append(i)
    elif methodList[i] == "Eclipse Timing Variations":
        eclipseTimingVariationsList.append(i)
    elif methodList[i] == "Imaging":
        imagingList.append(i)
    elif methodList[i] == "Microlensing":
        microlensingList.append(i)
    elif methodList[i] == "Orbital Brightness Modulation":
        orbitalBrightnessModulationList.append(i)
    elif methodList[i] == "Pulsar Timing":
        pulsarTimingList.append(i)
    elif methodList[i] == "Pulsation Timing Variations":
        pulsationTimingVariationsList.append(i)
    elif methodList[i] == "Radial Velocity":
        radialVelocityList.append(i)
    elif methodList[i] == "Transit":
        transitList.append(i)
    elif methodList[i] == "Transit Timing Variations":
        transitTimingVariationsList.append(i)

#Astrometry coordinates
x1, y1 = [], []
for i in astrometryList:
    x1.append(orbitalPeriodList[i])
    y1.append(massList[i])
    
#Eclipse Timing Variations coordinates
x2, y2 = [], []
for i in eclipseTimingVariationsList:
    x2.append(orbitalPeriodList[i])
    y2.append(massList[i])

#Imaging coordinates
x3, y3 = [], []
for i in imagingList:
    x3.append(orbitalPeriodList[i])
    y3.append(massList[i])
    
#Microlensing coordinates
x4, y4 = [], []
for i in microlensingList:
    x4.append(orbitalPeriodList[i])
    y4.append(massList[i])

#Orbital Brightness Modulation coordinates
x5, y5 = [], []
for i in orbitalBrightnessModulationList:
    x5.append(orbitalPeriodList[i])
    y5.append(massList[i])
    
#Pulsar Timing coordinates
x6, y6 = [], []
for i in pulsarTimingList:
    x6.append(orbitalPeriodList[i])
    y6.append(massList[i])
    
#Pulsation Timing Variations coordinates
x7, y7 = [], []
for i in pulsationTimingVariationsList:
    x7.append(orbitalPeriodList[i])
    y7.append(massList[i])
    
#Radial Velocity coordinates
x8, y8 = [], []
for i in radialVelocityList:
    x8.append(orbitalPeriodList[i])
    y8.append(massList[i])
    
#Transit coordinates
x9, y9 = [], []
for i in transitList:
    x9.append(orbitalPeriodList[i])
    y9.append(massList[i])

#Transit Timing Variations coordinates
x10, y10 = [], []
for i in transitTimingVariationsList:
    x10.append(orbitalPeriodList[i])
    y10.append(massList[i])


plt.figure()
plt.scatter(x1,y1, color = "red")
plt.scatter(x2,y2, color = "blue")
plt.scatter(x3,y3, color = "yellow")
plt.scatter(x4,y4, color = "orange")
plt.scatter(x5,y5, color = "purple")
plt.scatter(x6,y6, color = "green")
plt.scatter(x7,y7, color = "orange")
plt.scatter(x8,y8, color = "pink")
plt.scatter(x9,y9, color = "teal")
plt.scatter(x10,y10, color = "black")
plt.title("Relation Between Orbital Period and Planet Mass")
plt.xlabel("Orbital Period (Days)")
plt.ylabel("Planet Mass (In relation to Jupiter's Mass)")
plt.show()

#most of the planet mass data points are NaN, so I'm only left with a handful of data points
#which is there's so few points on the scatter plot :p
```
