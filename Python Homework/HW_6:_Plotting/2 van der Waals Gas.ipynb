```python
import numpy as np
import matplotlib.pyplot as plt

#x = Volume, y = Pressure
x,y = np.meshgrid(np.linspace(10,30,31),np.linspace(0,10,11))

#Temperature
z = ((y + 16.02*(1/(x**2)))*(x - 0.1124)/0.083144)

plt.figure()
plt.imshow(z, cmap= "coolwarm")
plt.xlabel("Volume (liters)")
plt.ylabel("Pressure (atm)")
plt.colorbar()
plt.show()
```
