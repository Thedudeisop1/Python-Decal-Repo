```python
import numpy as np
import matplotlib.pyplot as plt
import random

def randomNumbers():
    i = 40
    y = []
    while i > 0:
        j = random.randint(0,100)
        y.append(j)
        i -= 1
    return y

y1 = randomNumbers()
y2 = randomNumbers()
y3 = randomNumbers()
x = np.arange(0,40)

fig, axs = plt.subplots(2)
fig.suptitle("Random Data Vertically Stacked Subplots")
axs[0].plot(x,y1, color = "orange", linewidth = "10")
axs[0].plot(x,y2, color = "red", linestyle = "dashed")
axs[1].scatter(x,y3, color = "magenta", marker = "d")
plt.show()
```
