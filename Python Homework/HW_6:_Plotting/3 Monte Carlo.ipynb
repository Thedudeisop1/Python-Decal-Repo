```python
import numpy as np
import matplotlib.pyplot as plt
import random
import decimal

#Gets points for x
def PlotpointsX():
    xlist = []
    N = 1e4
    while N > 0:
        x = float(decimal.Decimal(random.randrange(0, 100000))/100000)
        xlist.append(x)
        N -= 1
    return xlist

#Gets points for y
def PlotpointsY():
    ylist = []
    N = 1e4
    while N > 0:
        y = float(decimal.Decimal(random.randrange(0, 100000))/100000)
        ylist.append(y)
        N -= 1
    return ylist

x = PlotpointsX()
y = PlotpointsY()

def pi(x,y):
    NInsidePi = 0
    for i in x:
        j = i
        k = y[x.index(i)]
        theta = np.arccos(j)
        p = j*np.tan(theta)
        if k < p:
            NInsidePi += 1
    pi = 4*(NInsidePi/10000)
    return pi

pi = pi(x,y)
print("Pi is approximately", pi)

def InsidePiY(x,y):
    list1 = []
    list2 = []
    for i in x:
        j = i
        k = y[x.index(i)]
        theta = np.arccos(j)
        p = j*np.tan(theta)
        if k < p:
            list1.append(i)
            list2.append(k)
    return list1, list2

x1,y1 = InsidePiY(x,y)

def OutsidePiY(x,y):
    list1 = []
    list2 = []
    for i in x:
        j = i
        k = y[x.index(i)]
        theta = np.arccos(j)
        p = j*np.tan(theta)
        if k > p:
            list1.append(i)
            list2.append(k)
    return list1, list2

x2,y2 = OutsidePiY(x,y)

plt.figure()
plt.title("Pi when N = 1e4")
plt.scatter(x1,y1, color= "red")
plt.scatter(x2,y2, color= "blue")
plt.show()
```
